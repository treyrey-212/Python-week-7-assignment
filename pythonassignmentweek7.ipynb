```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the red wine dataset
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
df_red = pd.read_csv(url, delimiter=';')

# Display the first few rows of the dataset
print(df_red.head())
```

```python
df = pd.read_csv(url, delimiter=';')
```

```python
# Check data types and missing values
print(df.info())

# Check for missing values
print(df.isnull().sum())

# Display basic statistics for the numerical columns
print(df.describe())
```

```python
# Drop rows with missing values
df_cleaned = df.dropna()

# Confirm there are no missing values
print(df_cleaned.isnull().sum())
```

```python
# Group by 'quality' and calculate the mean of 'alcohol'
avg_alcohol = df_cleaned.groupby('quality')['alcohol'].mean()

# Display the result
print(avg_alcohol)
```

```python
# 1. Line chart of average alcohol content by wine quality
avg_alcohol.plot(kind='line', color='blue', marker='o')
plt.title('Average Alcohol Content by Wine Quality')
plt.xlabel('Wine Quality')
plt.ylabel('Average Alcohol Content')
plt.grid(True)
plt.show()
```

```python
# 2. Bar chart of average alcohol content by wine quality
avg_alcohol.plot(kind='bar', color='lightcoral')
plt.title('Average Alcohol Content by Wine Quality')
plt.xlabel('Wine Quality')
plt.ylabel('Average Alcohol Content')
plt.xticks(rotation=0)
plt.show()
```

```python
# 3. Histogram of alcohol content
df_cleaned['alcohol'].plot(kind='hist', bins=20, color='green', edgecolor='black')
plt.title('Distribution of Alcohol Content')
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.show()
```

```python
# 4. Scatter plot of alcohol content vs residual sugar
df_cleaned.plot(kind='scatter', x='alcohol', y='residual sugar', color='purple')
plt.title('Alcohol Content vs Residual Sugar')
plt.xlabel('Alcohol Content')
plt.ylabel('Residual Sugar')
plt.show()
```
